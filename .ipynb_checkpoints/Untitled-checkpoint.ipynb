```python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.title("Sample Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
import matplotlib.pyplot as plt
x=['mon','tue','wed','thu','fri']
y=[10,6,21,17,20]
plt.plot(x,y,color='red',linestyle=':',linewidth=2,marker='o',label='2025 sales data')
plt.title("sales chart")
plt.legend(loc='upper left',fontsize=12)
plt.grid(color='black',linestyle='-',linewidth=1)
plt.xlabel("dayes of week")
plt.ylabel("sales per day")
plt.ylim(0,25)
plt.xlim(0,'fri')
plt.xticks(['mon','tue','wed','thu','fri'],['Day1','Day2','Day3','Day4','Day5'])
#plt.yticks([10,6,21,17,20],['quater1','quater2','quater3','quater4','quater5'])
plt.show()
```
