```python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.title("Sample Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python

import matplotlib.pyplot as plt
x=['mon','tue','wed','thu','fri']
y=[10,6,21,17,20]
plt.plot(x,y,color='brown',linestyle=':',linewidth=2,marker='o',label='2025 sales data')
plt.title("sales chart")
plt.legend(loc='upper left',fontsize=12)
plt.grid(color='black',linestyle='-',linewidth=1)
plt.xlabel("dayes of week")
plt.ylabel("sales per day")
plt.ylim(0,25)
plt.xlim(0,'fri')
plt.xticks(['mon','tue','wed','thu','fri'],['Day1','Day2','Day3','Day4','Day5'])
#plt.yticks([10,6,21,17,20],['quater1','quater2','quater3','quater4','quater5'])
plt.show()
```

```python
#plotting bar charts
import matplotlib.pyplot as plt
product=["A","B","C","D"]
sales=[1000,2100,800,1800]

plt.bar(product,sales,color='orange',linestyle="-",linewidth="2",label="sales data")
plt.xlabel("product")
plt.ylabel("product sales")
plt.title("product sales comparison")
plt.legend(loc="upper right",fontsize=12)
plt.show()
```

```python
#plotting pie chart 
import matplotlib.pyplot as plt
#plt.pie(values,labels_list,colurs,autopercent(%1.1f%%1)use to tell matplotlib to show percntage )
region=["north","south","east","west"]
revenue=[1000,1200,200,2000]
plt.pie(revenue,labels=region,autopct='%1.1f%%',colors=["gold","skyblue","lightgreen","cyan"])
plt.title="Revenue Contribution By Region"
plt.show()
```
