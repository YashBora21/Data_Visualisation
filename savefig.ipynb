```python
#saveplot(filename.extension,dpi value,bbox_inches='tight')
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,23,25,30]
#fig,ax=plt.subplot(nrow,ncols,figureSize=(width,heigth))
fig, ax=plt.subplots(1,2,figsize=(10,7))
ax[0].plot(x,y,color="red")
ax[0].set_title("line Plot")
ax[0].set_xlabel("X-axis")
ax[0].set_ylabel("Y-axis")
ax[1].bar(x,y,color="blue")
fig.suptitle("Comparison of line bar and line chart")
ax[1].set_xlabel("X-axis")
ax[1].set_ylabel("Y-axis")
ax[1].set_title("bar Plot")
plt.tight_layout()
plt.savefig('mat.png',dpi=360,bbox_inches='tight')
plt.show()
```
