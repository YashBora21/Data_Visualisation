```python
import matplotlib.pyplot as plt
hrs=[1,2,3,4,5,6,7,8]
score=[50,55,60,65,70,80,85,90]
plt.scatter(hrs,score,color="red",marker='o',label="Student data")
plt.xlabel("Hrs studied")
plt.ylabel("Exam Scores")
plt.title("Relation between Exam Score and HRS")
plt.grid(True)
plt.legend()
plt.show()
```
