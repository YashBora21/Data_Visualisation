```python
#plt.subplot(nrows,ncols,index(start fom zero))
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,23,25,30]
plt.subplot(1,2,1)#1 row 2nd column at 1 pos 
plt.plot(x,y)
plt.title("line chart")
plt.subplot(1,2,2)#1 row 2nd column at 2 pos 
plt.bar(x,y)
plt.title("bar chart")
plt.show()
```

```python
#this better approach 
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,23,25,30]
#fig,ax=plt.subplot(nrow,ncols,figureSize=(width,heigth))
fig, ax=plt.subplots(1,2,figsize=(10,7))
ax[0].plot(x,y,color="red")
ax[0].set_title("line Plot")
ax[1].bar(x,y,color="blue")
fig.suptitle("Comparison of line bar and line chart")
ax[1].set_title("bar Plot")
plt.tight_layout()
plt.show()

```
